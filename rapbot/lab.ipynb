```python
import requests
from bs4 import BeautifulSoup
```

```python
def read(url):
    result = requests.get(url)
    src = result.content
    soup = BeautifulSoup(src, 'html.parser')
    page = soup.find_all('div', {'class': None})
    
    lyrics = []
    
    for rows in page:
    	lyrics.append(rows.text)
    
    text = [word for line in lyrics for word in line.split()]

    return text
```
